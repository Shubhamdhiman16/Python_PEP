```python
#PART E : library management system store books  , add books  , search books  , issue , return  , create backup ,, sort books and display fimal library using appripriate list method 


library = ["Python", "Java", "C++"]
library.append("DBMS")
library.append("AI")
book = input("Enter book to search: ")
if book in library:
    print(book, "is available")
else:
    print(book, "is not available")

issue = input("Enter book to issue: ")
if issue in library:
    library.remove(issue)
    print(issue, "issued")
else:
    print("Book not available")

ret = input("Enter book to return: ")
library.append(ret)

backup = library.copy()

library.sort()

print("Final Library:", library)
print("Backup Library:", backup)
```
