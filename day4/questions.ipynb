```python
num = 5834
sum_digits = 0

for digit in str(num):
    sum_digits += int(digit)

print("Sum of digits", num, "=", sum_digits)
```

```python
# Print numbers from 1 to 20, skipping multiples of 3
for i in range(1, 21):
    if i % 3 == 0:
        continue  
    print(i)
```

```python
# Print numbers from 1 onwards, skipping multiples of 4, stopping when > 20
i = 1
while True:
    if i > 20:
        break 
        
    if i % 4 == 0:
        i += 1
        continue  
        
    print(i)
    i += 1
```

```python
#Palindrome
num = input("Enter a number: ")

if num == num[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
