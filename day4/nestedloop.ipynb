```python
#Q2
for i in range(1,4):
    print("Row" ,i,)
    for j in range(1,4):
        print("\tcolumn" ,j )
              
```

```python
# Outer loop for 3 rows
for row in range(3):
   
    for col in range(4):
        print("*", end=" ")
    print()  
```

```python
# Outer loop for 3 rows
for row in range(3):
   
    for col in range(1, 5):
        print(col, end=" ")
    print() 
```

```python
# Outer loop for 3 rows
for row in range(3):
   
    for col in range(65, 69):
        print(chr(col), end=" ")
    print() 
```

```python
# Q8
num = 1

for row in range(3):
    
    for col in range(3):
        print(num, end=" ")
        num += 1  
    print()  # New line after each row
```
