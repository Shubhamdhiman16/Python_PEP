```python
movies = ["KGF", "Pushpa", "Leo", "Jawan", "Animal", "Kalki"]

print("Original List:", movies)
print("First Item:", movies[0])
print("Last Item:", movies[-1])
print("Top 3 Items:", movies[:3])
print("Last 3 Items:", movies[-3:])
print("Alternate Items:", movies[::2])
print("Reverse List:", movies[::-1])
```

```python
heart_rates = [72, 105, 98, 120, 85, 110, 95]
count = 0
for rate in heart_rates:
    if rate > 100:
        print(rate, "BPM")
        count += 1

print("Total Patients:", count)
```

```python
salary_list = [45000, 72000, 32000, 85000, 60000, 55000]
highest_salary = max(salary_list)

lowest_salary = min(salary_list)

total_salary = sum(salary_list)

average_salary = sum(salary_list) / len(salary_list)

print(f"Highest Salary: {highest_salary:,}")
print(f"Lowest Salary:  {lowest_salary:,}")
print(f"Total Salary:   {total_salary:,}")
print(f"Average Salary: {average_salary:,.2f}")
```
