```python
text = input("Enter a string: ")

letters = digits = spaces = special = 0

for ch in text:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1
    else:
        special += 1

print("Letters =", letters)
print("Digits =", digits)
print("Spaces =", spaces)
print("Special Characters =", special)
```

```python
str1 = input("Enter first string: ").lower()
str2 = input("Enter second string: ").lower()

if sorted(str1) == sorted(str2):
    print("Anagram")
else:
    print("Not Anagram")
```
