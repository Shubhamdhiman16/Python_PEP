```python
text = input("Enter a string: ")

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

```python
playlist1 = {"Navior", "checkin", "sign to god", "champion"}
playlist2 = {"headliner", "Navior", "Closer", "sign to god"}

print("Common Songs:", playlist1 & playlist2)
print("Only in Playlist 1:", playlist1 - playlist2)
print("Only in Playlist 2:", playlist2 - playlist1)
```
