```python
# Global variable
balance = 1000

def deposit(amount):
    global balance
    balance += amount
    print("Deposited:", amount)
    print("Current Balance:", balance)

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
        print("Withdrawn:", amount)
        print("Current Balance:", balance)
    else:
        print("Insufficient Balance!")

deposit(500)
withdraw(300)
withdraw(1500)

```
