```python
square = lambda x: x * x

print(square(5))
```

```python
add = lambda a, b: a + b

print(add(10, 20))
```

```python
largest = lambda a, b: a if a > b else b

print(largest(10, 20))
```

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
numbers = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```

```python
check = lambda n: "Positive" if n > 0 else ("Negative" if n < 0 else "Zero")

print(check(10))
print(check(-5))
print(check(0))
```

```python
words = ["apple", "banana", "mango", "watermelon"]

longest = lambda x: max(x, key=len)

print(longest(words))
```

```python
discount = lambda price: price - (price * 20 / 100)

print(discount(1000))
```

```python
n = int(input("Enter a number: "))

numbers = list(map(lambda x: n + x, range(4)))

print(numbers)
```

```python
names = ["shubham", "rahul", "dhiman"]

upper_names = list(map(lambda x: x.upper(), names))

print(upper_names)
```
