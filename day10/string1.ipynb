```python
text = input("Enter a string: ").lower()

count = (text.count('a') + text.count('e') +
         text.count('i') + text.count('o') +
         text.count('u'))

print("Number of vowels =", count)
```

```python
text = input("Enter a string: ").lower()

if text == "".join(reversed(text)):
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
text = input("Enter a string: ")

count = 0

for ch in text:
    if ('a' <= ch <= 'z') or ('A' <= ch <= 'Z'):
        count += 1

print("Number of alphabets:", count)
```

```python
email = input("Enter email: ")

if email.count("@") == 1 and "." in email.split("@")[1]:
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
phone = input("Enter phone number: ")

masked = "*" * (len(phone) - 4) + phone[-4:]

print(masked)
```

```python
text = input("Enter a string: ")
word = input("Enter the word: ")

first = text.find(word)
second = text.find(word, first + 1)

print("Second occurrence index:", second)
```
