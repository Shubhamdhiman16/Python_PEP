```python
# create a student class with roll_no , name , marks . add method get_grade() thatreturns "A" if marks >= 90 "||" if >. 75 else "C".

class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks

    def get_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"

s = Student(101, "Shubham", 88)

print("Roll No:", s.roll_no)
print("Name:", s.name)
print("Marks:", s.marks)
print("Grade:", s.get_grade())
```

```python
# create a rectangle class with length and width . add methods area() and perimeter()


class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)

r = Rectangle(10, 5)

print("Area:", r.area())
print("Perimeter:", r.perimeter())
```

```python
#create an employee class with name and salary . add a method give_raise(percent) that updates seld.salery in place (no return , just modify and print new salary)

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def give_raise(self, percent):
        self.salary += self.salary * percent / 100
        print("New Salary:", self.salary)

e = Employee("Shubham", 50000)

e.give_raise(10)
```

```python
# create a library class that holds a list of book object . add method add_book(book) and show_all_books()


class Book:
    def __init__(self , title):
        self.title = title

class Library:
    def __init__(self):
        self.books = []
    def add_book(self , book):
        self.books.append(book)
    def show_add_books(self):
        for book in self.books:
            print(book.title)

b1 = Book("Pyhton")
b2 = Book("Java")
lib = Library()
lib.add_book(b1)
lib.add_book(b2)
lib.show_add_books()
```

```python
# create an employee class with a private attribute __salary. add a getter and setter.  the setter should rejectc negative salaries . Add method apply_raise(percent) that increase salary using setter internally.


class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary >= 0:
            self.__salary = salary
        else:
            print("Invalid salary")
        
    def apply_raise(self, percent):
        new_salary = self.__salary + (self.__salary * percent / 100)
        self.set_salary(new_salary)
        print("New Salary:", self.__salary)

e = Employee("Shubham", 50000)

print("Salary:", e.get_salary())
e.apply_raise(10)
```
