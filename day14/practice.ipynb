```python
#given a sentence display the frequecy of every character


sentence = input("Enter a sentence: ")
freq = {}
for ch in sentence:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)
```

```python
#maintain a simple contact directory supporting search by name
contacts = {
    "Shubham": "9876543210",
    "Aman": "9876501234",
    "Rahul": "9876511111",
    "Simran": "9888822234"
}

name = input("Enter name to search: ")

if name in contacts:
    print("Phone Number:", contacts[name])
else:
    print("Contact not found")
```

```python
#given marks of students , print names of students scoring above class average
marks = {
    "Shubham": 85,
    "Aman": 70,
    "Rahul": 90
}

avg = sum(marks.values()) / len(marks)

for name in marks:
    if marks[name] > avg:
        print(name)
```

```python
seats = ["A1", "A2", "A3", "A2", "A4", "A5", "A1"]
booked = set()
duplicate = set()
for seat in seats:
    if seat in booked:
        duplicate.add(seat)
    else:
        booked.add(seat)
print("Double Booked Seats:", duplicate)
```
