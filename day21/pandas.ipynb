```python
import pandas as pd

data = [
    ["Aman", 20, 85],
    ["Riya", 21, 90],
    ["Karan", 22, 78]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])

print(df)

df.head(1)
df.sample(2)
df.describe()
df.shape
df.columns
df.index
df.types
df('col').value_counts()
df('col').unique()
```
