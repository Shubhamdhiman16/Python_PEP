```python
#factorial using recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter a number: "))
print("Factorial =", factorial(num))
```

```python
#Sum of n number using recursion
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n - 1)

num = int(input("Enter a number: "))
print("Sum =", sum_n(num))
```

```python
#recursion 
def list_sum(lst):
    if len(lst) == 0:      
        return 0
    return lst[0] + list_sum(lst[1:])   

numbers = [12, 5 , 7, 10 , 8]

print("Sum =", list_sum(numbers))
```

```python
def calculator(operation):
    if operation == "+":
        return lambda a, b: a + b
    elif operation == "-":
        return lambda a, b: a - b
    elif operation == "*":
        return lambda a, b: a * b
    elif operation == "/":
        return lambda a, b: a / b

calc = calculator("+")
print(calc(5,3))
```

```python
def my_map(func, data):
    result = []
    for i in data:
        result.append(func(i))
    return result

numbers = [1, 2, 3, 4]

print(my_map(lambda x: x * 2, numbers))
```

```python
def compress(s, i=0, count=1):
    if i == len(s) - 1:
        return s[i] + str(count)

    if s[i] == s[i + 1]:
        return compress(s, i + 1, count + 1)

    return s[i] + str(count) + compress(s, i + 1, 1)

s = "aaaabbbbbccccddddaaa"
print(compress(s))
```

```python
words = ["Zero", "One", "Two", "Three", "Four",
         "Five", "Six", "Seven", "Eight", "Nine"]

def digit_to_words(n):
    if n < 10:
        print(words[n], end=" ")
        return
    digit_to_words(n // 10)
    print(words[n % 10], end=" ")

num = int(input("Enter a number: "))
digit_to_words(num)
```
