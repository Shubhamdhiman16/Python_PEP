```python
#project student marks analysis system
#this walkthorugh a smalll , realistic anylisis pipeline using only numpy . have praticipates type each along with you  pause after each blocks to check output
#step 1: dataset creation
#step2 : average marks per student
#step3 : highest and lowest marks
#step4: standard deviation per student
#step5 : ranking students 
#stdep6: top tudents 
#step 7: subject wise analysis
#stuep8: percentage calculation
#step9: visulization data

#the final step reshapes result into simple (lable , value) pairs to hand to matplotilab or pandas 


import numpy as np
import matplotlib.pyplot as plt

# Step 1: Dataset Creation
students = np.array(["Aman", "Riya", "Neha", "Rahul", "Priya"])

marks = np.array([
    [85, 90, 88],   # Aman
    [78, 82, 80],   # Riya
    [92, 95, 91],   # Neha
    [70, 68, 72],   # Rahul
    [88, 84, 90]    # Priya
])

subjects = ["Math", "Science", "English"]

print("Marks Matrix:\n", marks)

# Step 2: Average Marks per Student
average = np.mean(marks, axis=1)
print("\nAverage Marks:")
for i in range(len(students)):
    print(students[i], ":", round(average[i], 2))

# Step 3: Highest and Lowest Marks
highest = np.max(marks, axis=1)
lowest = np.min(marks, axis=1)

print("\nHighest & Lowest Marks:")
for i in range(len(students)):
    print(students[i], "Highest =", highest[i], "Lowest =", lowest[i])

# Step 4: Standard Deviation
std = np.std(marks, axis=1)

print("\nStandard Deviation:")
for i in range(len(students)):
    print(students[i], ":", round(std[i], 2))

# Step 5: Ranking Students
total = np.sum(marks, axis=1)
rank = np.argsort(total)[::-1]

print("\nStudent Ranking:")
for i in range(len(rank)):
    print(i + 1, ".", students[rank[i]], "-", total[rank[i]])

# Step 6: Top Student
top = np.argmax(total)
print("\nTop Student:", students[top], "-", total[top])

# Step 7: Subject-wise Analysis
print("\nSubject-wise Average:")
for i in range(len(subjects)):
    print(subjects[i], ":", round(np.mean(marks[:, i]), 2))

# Step 8: Percentage Calculation
percentage = total / 300 * 100

print("\nPercentage:")
for i in range(len(students)):
    print(students[i], ":", round(percentage[i], 2), "%")

# Step 9: Visualization
plt.bar(students, percentage)
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Student Marks Analysis System")
plt.show()
```
